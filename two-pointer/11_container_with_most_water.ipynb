You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container.

 ![alt text](image.png)

Example 1:


Input: height = [1,8,6,2,5,4,8,3,7]
Output: 49
Explanation: The above vertical lines are represented by array [1,8,6,2,5,4,8,3,7]. In this case, the max area of water (blue section) the container can contain is 49.
Example 2:

Input: height = [1,1]
Output: 1

```python
from typing import List
class Solution:
    def maxArea(self, height: List[int]) -> int:
        length = len(height)
        left = 0
        right = length - 1
        max_area = 0
        while left < right:
            area = min(height[left], height[right]) * (right - left)
            max_area = max(max_area, area)
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
        return max_area
```

Above code works fine but this takes more time as we are using built-in min and max function of Python.

Below code snippet is optimized by removing min and max

```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        length = len(height)
        left = 0
        right = length - 1
        max_area = 0
        while left < right:
            if height[left] < height[right]:
                minvalue = height[left]
            else:
                minvalue = height[right]
            area = minvalue * (right - left)
            max_area = max_area if max_area > area else area
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
        return max_area
```
