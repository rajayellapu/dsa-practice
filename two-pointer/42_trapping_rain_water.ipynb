Given n non-negative integers representing an elevation map where the width of each bar is 1, compute how much water it can trap after raining.

 

Example 1:


Input: height = [0,1,0,2,1,0,1,3,2,1,2,1]
Output: 6
Explanation: The above elevation map (black section) is represented by array [0,1,0,2,1,0,1,3,2,1,2,1]. In this case, 6 units of rain water (blue section) are being trapped.
Example 2:

Input: height = [4,2,0,3,2,5]
Output: 9

```python
from typing import List

class Solution:
    def mydef(self, height:List[int]) -> int:
        left = 0
        right = len(height) - 1
        temp = 0
        watercount = 0
        while left < right:
            if(left == 0 and right == len(height) - 1):
                temp = height[left] if height[left] < height[right] else height[right]

            if(height[left] < height[right]):
                if(temp < height[left]):
                    temp = height[left] if height[left] < height[right] else height[right]
                watercount = temp - height[left]
                left += 1
            else:
                if(temp < height[right]):
                    temp = height[right] if height[right] < height[left] else height[left]
                watercount = temp - height[left]
                right -= 1
        return watercount      
```

This is two pointer pattern. one index starts from first element and other from last element of the array. Then we have a temp variable which holds the min of both values.

WaterCounter is counting numbers of blocks which contain water.
